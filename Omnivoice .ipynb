# **Change in App File**

```python
# Update app.py launch line to add share=True

file_path = "app.py"   # change path if app.py is in another folder

with open(file_path, "r", encoding="utf-8") as f:
    content = f.read()

content = content.replace(
    "demo.queue().launch()",
    "demo.queue().launch(share=True)"
)

with open(file_path, "w", encoding="utf-8") as f:
    f.write(content)

print("✅ app.py updated successfully with share=True")
```

```python
file_path = "/content/OmniVoice/omnivoice/models/omnivoice.py"

with open(file_path, "r") as f:
    code = f.read()

old = 'return self._asr_pipe(audio_input)["text"].strip()'

new = '''import numpy as np
        if isinstance(audio_input, dict):
            audio_array = np.array(audio_input["array"], dtype=np.float32)
            sr = audio_input["sampling_rate"]
        else:
            audio_array = np.array(audio_input[0], dtype=np.float32)
            sr = audio_input[1]
        return self._asr_pipe({"array": audio_array, "sampling_rate": sr})["text"].strip()'''

code = code.replace(old, new)

with open(file_path, "w") as f:
    f.write(code)

print("Patch applied correctly ✅")
```
